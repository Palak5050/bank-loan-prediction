```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings(action="ignore")
```

```python
df=pd.read_csv("bank_data.csv",sep=';')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(["y"],axis=1)
y=df.y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=32)
```

```python
from catboost import CatBoostClassifier
```

```python
from catboost import Pool
```

```python
pool_train=Pool(X_train,y_train,cat_features=["job","marital","education","default","housing","loan","contact","month","day_of_week","duration","poutcome"])
pool_test=Pool(X_test,y_test,cat_features=["job","marital","education","default","housing","loan","contact","month","day_of_week","duration","poutcome"])	
```

```python
cbc=CatBoostClassifier(iterations=100)
```

```python
cbc.fit(pool_train)
```

```python
cbc.score(pool_train)
```

```python
cbc.score(pool_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred=cbc.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
```
